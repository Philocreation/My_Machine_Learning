```python
 import numpy as np
 import pandas as pd
 from sklearn.tree import DecisionTreeClassifier
 from sklearn.metrics import accuracy_score
 from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('/content/drive/MyDrive/Fish.csv')
df
```

```python
print(df.info())
```

```python
df_1 = df.iloc[:,1:]
print(df_1.info())
df_1.reset_index(drop=True)
```

```python
from sklearn.linear_model import LinearRegression

features = df_1.iloc[:,2:].values
label = df_1.iloc[:,2:].values
lr = LinearRegression()

kfold = KFold(n_splits = 3)
cv_accuracy = []
print('물고기 데이터 세트 크기:', features.shape[0])
```

```python
n_iter = 0

#KFold 객체의 split()를 호출하면 폴드 별 학습용, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
  #kfold.split()으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
  x_train, x_test = features[train_index], features[test_index]
  y_train, y_test = label[train_index], label[test_index]
  #학습 및 예측
  lr.fit(x_train, y_train)
  pred = lr.predict(y_test)
  n_iter += 1
  #반복 시마다 정확도 측정
  
  accuracy = np.round(lr.score(x_test,pred),4)
  train_size = x_train.shape[0]
  test_size = x_test.shape[0]
  print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(n_iter, test_index))
  cv_accuracy.append(accuracy)

#개별 iteraion별 정확도를 합하여 평균 정확도 계산
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```
